2. Operaciones matematicas y de sintesis

Pandas contiene varios metodos que usan operaciones matematicas para resumir una columna o una fila. Estos metodos funcionan unicamente en las columanas con datos numericos.

Los ,as usados son:

* min() y max(): calcual el minimo /maximo de cada fila o columna
* sum() : calcual la suma de los elementos de cada fila o columna
* mean(), std, median():calcula la media /desviacion estandar /mediana de cada fila o columna.

Por defecto estas operaciones se aplican sobre las columnas (axis = 1)pero si definimos axis = 0 el calculo se hara por filas:

Haz doble clic (o pulsa Intro)para editar


```python
#calcular el minimo de cada columna  RESULTADO DE LA EXPLICACION ANTERIOR
df.min()
```

```python
df.select_dtypes(include='number').min() # y probar con max(), sum(), mean(), std()
```

```python
df.select_dtypes(include='number').max() # y probar con max(), sum(), mean(), std()
```

```python
df.select_dtypes(include='number').sum() # y probar con max(), sum(), mean(), std()
```

La forma más sencilla de sumar todos los valores de un dataframe de Pandas es usar el método sum() . Este método calcula la suma de cada columna del dataframe y devuelve un nuevo dataframe con los resultados

```python
df.select_dtypes(include='number').mean() # y probar con max(), sum(), mean(), std()
```

Mean: devuelve un valor numérico que se aproxima al promedio aritmético de sus argumentos. El tipo de función es numérico.

```python
df.select_dtypes(include='number').std() # y probar con max(), sum(), mean(), std()
```

la función std() en pandas obtiene la desviación estándar de los valores de un eje especificado de un DataFrame dado. Matemáticamente, La desviación estándar se define como la medida de la dispersión de cada valor de un conjunto de datos con respecto a la media.
